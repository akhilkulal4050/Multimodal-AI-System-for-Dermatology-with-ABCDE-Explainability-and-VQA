# Hugging Face CLI Setup and Login Steps

1. First, ensure you have a Hugging Face account. If not, create one at [Hugging Face](https://huggingface.co/join).

2. Install the Hugging Face CLI (which you already have in your environment)

3. Login to Hugging Face using the CLI:
   - Open terminal
   - Run `huggingface-cli login`
   - Enter your token when prompted (you can find your token at https://huggingface.co/settings/tokens)
   - The token will be saved locally

4. Verify your login:
   - You can run `huggingface-cli whoami` to verify your login status

Note: Your authentication token will be stored in `~/.huggingface/token` and will be used automatically by the datasets library.

```python
from datasets import load_dataset

# Login using e.g. `huggingface-cli login` to access this dataset
ds = load_dataset("redlessone/Derm1M")
print(ds)
```

```python
print(ds['filename'][0])
```

```python
# Convert the dataset to pandas DataFrame
import pandas as pd

# Convert the dataset to pandas DataFrame
df = pd.DataFrame(ds['validation'])  # assuming we want to work with the validation split
print("Dataset shape:", df.shape)
print("\nFirst few rows of the dataset:")
print(df.head())
```

```python
# Save the DataFrame to a CSV file
csv_filename = 'derm1m_dataset.csv'
df.to_csv(csv_filename, index=False)
print(f"Dataset saved to {csv_filename}")

# Display the first few rows of the saved CSV file to verify
print("\nVerifying the saved CSV file:")
print(pd.read_csv(csv_filename).head())
```
